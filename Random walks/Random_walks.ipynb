This will be a short notebook on how I will try to generate/simulate (uniform) random walks in Python.

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
# Probability for success and failures
p = 0.5
q = 1-p

# Time steps and paths
steps = 30
paths = 100
```

```python
# Generate matrix of uniform sampling
A = np.zeros((steps, paths+1))
for i in range(1,paths):
    A[:,i] = stats.uniform.rvs(size=steps)
# Matrix for direction
dir = np.sign(p-A[:,1:paths+1])

# Random walk proper
for t in range(paths):
    A[:,t+1] = A[:,t] + dir[:,t]

# Plotting
plt.figure(figsize=(10,6))
for t in range(paths-1):
    plt.plot(range(steps), A[:,t+1], "-")
```
